```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import chardet

outhome = 'C:/Users/70794/SDC/SDC_W3/outofhome.csv'


education = 'C:/Users/70794/SDC/SDC_W3/education.csv'

social = 'C:/Users/70794/SDC/SDC_W3/socialmedia.csv'

def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

df_out = pd.read_csv(outhome, encoding=detect_encoding(outhome))
df_edu = pd.read_csv(education, encoding=detect_encoding(education))
df_social = pd.read_csv(social, encoding=detect_encoding(social))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_edu = {
    'Education level': ['Hbo-, wo-master, doctor', 'Hbo-, wo-bachelor', 'Mbo 2,3,4, havo, vwo', 'Vmbo, avo onderbouw, mbo 1', 'Basic Education'],
    'Dutch (%)': [82.3, 82.9, 78.4, 70.0, 60.7],
    'Dialect (%)': [2.8, 4.4, 5.5, 7.1, 6.6],
    'Low Saxon (%)': [1.1, 1.7, 5.4, 7.8, 8.0],
    'Frisian (%)': [0.9, 1.8, 1.7, 3.2, 2.4],
    'Limburgish (%)': [1.9, 3.0, 3.6, 4.4, 4.1],
    'Other Language (%)': [10.9, 6.2, 5.4, 7.5, 18.2]
}

df_edu = pd.DataFrame(data_edu)

percentage_columns_edu = ['Dutch (%)', 'Dialect (%)', 'Low Saxon (%)', 'Frisian (%)', 'Limburgish (%)', 'Other Language (%)']
df_edu[percentage_columns_edu] = df_edu[percentage_columns_edu].replace(',', '.', regex=True).astype(float)

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(data=df_edu.melt(id_vars='Education level', var_name='Language', value_name='Percentage'),
            x='Education level', y='Percentage', hue='Language', palette='viridis')

plt.title('Language Usage by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Percentage (%)')

plt.legend(loc='upper right')

plt.xticks(rotation=45, ha='right')

plt.show()
```
